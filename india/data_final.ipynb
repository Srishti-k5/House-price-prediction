```python
import pandas as pd
```

```python
data = pd.read_excel("dataset.xlsx")
data
```

#### Separating Info column into BHK, Type and Location

```python
data['BHK'] = data['Info'].str.split('BHK').str.get(0)
data['Info1'] = data['Info'].str.split('BHK').str.get(1)
data = data.drop('Info', axis=1)
```

```python
data
```

```python
data['Type'] = data['Info1'].str.split('for').str.get(0)
data['Info2'] = data['Info1'].str.split('for').str.get(1)
data = data.drop('Info1', axis=1)
```

```python
data
```

```python
data['Location'] = data['Info2'].str.split().str.get(-1)
data = data.drop('Info2', axis=1)
```

```python
data
```

#### Separating Floor column into Total floor and floor on sale

```python
data['Floor Sale'] = data['Floor'].str.split().str.get(0)
data['Total Floor'] = data['Floor'].str.split().str.get(-1)
data = data.drop('Floor', axis=1)
```

```python
data
```

#### Creating function for replace some Status which are underconstruction

```python
# Function to replace some status with underconstruction
def replace(value):
    if value != "Ready to Move":
        return "Underconstruction"
    return value

# Apply the custom function to the specified column
data['Status'] = data['Status'].apply(replace)
```

```python
data
```

```python
data = data.drop('Unnamed: 0', axis=1)
```

```python
data
```

```python
# Rearrange columns
data = data[['BHK', 'Type', 'Location', 'Super Area', 'Carpet Area', 'Transaction', 'Status', 'Floor Sale', 'Total Floor',
            'Furnishing', 'Bathroom', 'Facing', 'Balcony', 'Prices']]
```

#### Both super area and carpet area is in sqft

```python
data
```

```python
data['Super Area'] = data['Super Area'].str.replace(' sqft', '')
data['Carpet Area'] = data['Carpet Area'].str.replace(' sqft', '')
```

```python
data
```

#### All Prices are in INR
#### Converting Lac to Crore

```python
data['Prices'] = data['Prices'].str.replace('₹', '')
```

```python
data['Prices']
```

```python
def convert_to_crores(value):
    if 'Lac' in value:
        numeric_value = float(value.replace(' Lac', '')) / 100.0
        return f'{numeric_value:.3f} Cr'
    else:
        return value

# Apply the custom function to the 'Value' column
data['Prices'] = data['Prices'].apply(convert_to_crores)
```

```python
data['Prices']
```

```python
data['Prices'] = data['Prices'].replace(' Cr', '')
```

```python
data
```
