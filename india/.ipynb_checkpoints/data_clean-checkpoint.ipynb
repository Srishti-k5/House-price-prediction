#### For conversion all units in area to sqft

```python
def convert_to_sqft(value):
    split_value = str(value).split(" ")
    if len(split_value) == 2:
        split_value[0] = split_value[0].replace(",","")
        if split_value[1] == 'sqyrd':
            split_value[0] = int(split_value[0])*9
            split_value.pop()
            split_value = split_value[0]
            return split_value
        
        if split_value[1] == 'maria':
            split_value[0] = int(split_value[0])*272.25
            split_value.pop()
            split_value = split_value[0]
            return split_value
        
        if split_value[1] == 'kottah':
            split_value[0] = int(split_value[0])*1361
            split_value.pop()
            split_value = split_value[0]
            return split_value
        
        if split_value[1] == 'sqm':
            split_value[0] = int(split_value[0])*10.76
            split_value.pop()
            split_value = split_value[0]
            return split_value
        
        if split_value[1] == 'acre':
            split_value[0] = int(split_value[0])*43560
            split_value.pop()
            split_value = split_value[0]
            return split_value
        
        if split_value[1] == 'cent':
            split_value[0] = int(split_value[0])*435.56
            split_value.pop()
            split_value = split_value[0]
            return split_value
        
        if split_value[1] == 'guntha':
            split_value[0] = int(split_value[0])*1089
            split_value.pop()
            split_value = split_value[0]
            return split_value
        
        if split_value[1] == 'aankadam':
            split_value[0] = int(split_value[0])*72
            split_value.pop()
            split_value = split_value[0]
            return split_value
        
        if split_value[1] == 'bigha':
            split_value[0] = int(split_value[0])*27000
            split_value.pop()
            split_value = split_value[0]
            return split_value
        
        if split_value[1] == 'kanal':
            split_value[0] = int(split_value[0])*5445
            split_value.pop()
            split_value = split_value[0]
            return split_value

        
    else:
        return value


# convert_to_sqft(x)
```

```python
data['Super Area'] = data['Super Area'].apply(convert_to_sqft)
data['Super Area'].unique()
```

```python
data['Carpet Area'].unique()
```

```python
data['Carpet Area'] = data['Carpet Area'].apply(convert_to_sqft)
data['Carpet Area'].unique()
```

```python
data['Super Area'] = pd.to_numeric(data['Super Area'], errors='coerce')
data['Carpet Area'] = pd.to_numeric(data['Carpet Area'], errors='coerce')
```

```python
data['Super Area'].fillna(data['Carpet Area'] * 1.43, inplace=True)
```

```python
data['Carpet Area'].fillna((data['Super Area'] / 1.43).round(2), inplace=True)
```

#### Still few none values so put average there

```python
average_value_sup = data['Super Area'].mean()
data['Super Area'].fillna(average_value_sup, inplace=True)
```

```python
average_value_car = data['Carpet Area'].mean()
data['Carpet Area'].fillna(average_value_car, inplace=True)
```

#### Generated random values for numerical columns considering std and for string columns considered proportions
